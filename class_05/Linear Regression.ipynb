```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_samples=200,n_features=1,noise=25.5,n_informative=1,random_state=1)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
print(X.shape)
print(Y.shape)
```

```python
def hypothesis(x,theta):
    return theta[0] + theta[1]*x

def predictions(X,theta):
    y_pred = []
    
    for i in range(X.shape[0]):
        pred = hypothesis(X[i],theta)
        y_pred.append(pred)
    y_pred = np.array(y_pred)
    return y_pred
    
def get_error(X,Y,theta):
    e = 0
    m = X.shape[0]
    for i in range(m):
        e += (hypothesis(X[i],theta)-Y[i])**2
        
    return e

def getGradients(X,Y,theta):
    
    grads = np.zeros((2))
    m = X.shape[0]
    
    for i in range(m):
        YPi = hypothesis(X[i],theta)
        grads[0] += (YPi - Y[i])
        grads[1] += (YPi - Y[i])*X[i]
        
    return grads
    
def gradientDescent(X,Y,maxItr=100,learning_rate=0.001):
    
    theta = np.zeros((2))
    error = []
    
    for i in range(maxItr):
        grad = getGradients(X,Y,theta)
        e = get_error(X,Y,theta)
        theta[0]= theta[0] - learning_rate*grad[0]
        theta[1]= theta[1] - learning_rate*grad[1]
        
        if(i%3==0):
            plt.figure(1)
            plt.scatter(X,Y)
            plt.plot(X,predictions(X,theta),c='red')
            plt.pause(2)
        
        error.append(e)
        
    return theta,error
```

```python
theta = np.zeros((2))
```

```python
print(theta)
```

```python
x = np.array([1,2,3])
np.sum(x)-1
```

```python
theta,error = gradientDescent(X,Y,maxItr=30)
```

```python
plt.plot(error)
plt.show()
```

```python
plt.scatter(X,Y)

x_test = np.linspace(-3,3,10)
y_test = []
for x_ in x_test:
    y_ = hypothesis(x_,theta)
    y_test.append(y_)

y_test = np.array(y_test)

plt.plot(x_test,y_test,c='orange')
plt.show()
```

```python
plt.scatter(X,Y)
plt.plot(x_test,y_pred,c='red')
plt.show()
```
