```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread("/home/vasu/Downloads/759351.jpg")
img = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)

print img.shape
plt.imshow(img, cmap='gray')
```

```python
# detects edges

kernel = np.array([[-1.0,-1.0,-1.0],
          [-1.0,8.0,-1.0],
          [-1.0,-1.0,-1.0]])

print kernel
```

```python
new_img = np.zeros((img.shape[0]-kernel.shape[0]+1,img.shape[1]-kernel.shape[1]+1))
```

```python
# convolution

for ix in range(new_img.shape[0]):
    for ij in range(new_img.shape[1]):
        
        #patch
        img_patch = img[ix:ix+kernel.shape[0], ij:ij+kernel.shape[1]]
        k_prod = img_patch*kernel
        
        new_img[ix][ij] = max(0,k_prod.sum())
```
